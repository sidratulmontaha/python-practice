```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(50)
Y = 2 * X + np.random.normal(0,0.1,50)

np.cov(X, Y)

np.corrcoef(X, Y)
```

```python
X = np.random.rand(50)
Y = np.random.rand(50)

plt.scatter(X, Y)
plt.xlabel("X value")
plt.ylabel("Y value")

print(np.cov(X, Y)[0,1] / (np.std(X) * np.std(Y)))
print(np.corrcoef(X, Y)[0,1])
```

```python
X = np.random.rand(50)
Y = X + np.random.normal(0,0.1,50)

plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')

print(np.corrcoef(X, Y)[0,1])
```

```python
X = np.random.rand(50)
Y = X + np.random.normal(0,0.2,50)

plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')

print(np.corrcoef(X,Y)[0,1])
```

```python
X = np.random.rand(50)
Y = -X + np.random.normal(0,0.1,50)

plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')

print(np.corrcoef(X,Y)[0,1])
```

```python
from datetime import datetime
import pandas_datareader as pdr
bench =  pdr.data.DataReader('SPY', data_source='yahoo', start=datetime(2013,1,1), end=datetime(2015,1,1)).iloc[:, -1]
a1 =  pdr.data.DataReader('LRCX', data_source='yahoo', start=datetime(2013,1,1), end=datetime(2015,1,1)).iloc[:, -1]
a2 =  pdr.data.DataReader('AAPL', data_source='yahoo', start=datetime(2013,1,1), end=datetime(2015,1,1)).iloc[:, -1]

plt.scatter(a1,a2)
plt.xlabel('LRCX')
plt.ylabel('AAPL')

print(np.corrcoef(a1,a2)[0,1])
print(np.corrcoef(a1,bench)[0,1])
print(np.corrcoef(a2,bench)[0,1])
```

```python
rolling_correlation = a1.rolling(60).corr(a2)

plt.plot(rolling_correlation)
```
