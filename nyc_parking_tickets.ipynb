```python
import collections
import csv
import datetime

class ParkingTicketIterator:
    def __init__(self):
        self._f = open('nyc_parking_tickets_extract.csv')
        self._csv_reader = iter(csv.reader(self._f, delimiter=','))
        self._csv_reader.__next__() #skip first row
        self._Ticket = collections.namedtuple("Ticket", "SummonsNumber,PlateID,RegistrationState,PlateType,IssueDate,ViolationCode,VehicleBodyType,VehicleMake,ViolationDescription")
    
    def __iter__(self):
        return self
    
    def __next__(self):
        row = self._csv_reader.__next__()
        print(row)
        if row is None:
            self._f.close()
            raise StopIteration
        else:
            t = self._Ticket(SummonsNumber=int(row[0]), PlateID=row[1], RegistrationState=row[2], PlateType=row[3], IssueDate=datetime.datetime.strptime(row[4], '%m/%d/%Y').date(), ViolationCode=row[5], VehicleBodyType=row[6], VehicleMake=row[7], ViolationDescription=row[8])
            return t
        
        
```

```python
p = ParkingTicketIterator()

for i in p:
    print(i)
```
