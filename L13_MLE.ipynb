```python
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats
```

```python
true_mean=40
true_std = 10

X = np.random.normal(true_mean, true_std, 1000)
X
```

```python
print(np.mean(X))
print(np.std(X))
```

```python
mu, std = scipy.stats.norm.fit(X)
print(mu, std)
```

```python
x = np.linspace(0,80,80)
plt.hist(X, bins = x, density=True)

plt.plot(scipy.stats.norm.pdf(x, loc=mu, scale=std))
```

```python
true_lambda = 5
#1/lambda * e^(-x/lambda)
X = np.random.exponential(true_lambda, 1000)
```

```python
_,l = scipy.stats.expon.fit(X, floc=0)

plt.hist(X, bins = range(0,80), density=True)

plt.plot(scipy.stats.expon.pdf(x, scale=l))
```
